```python
with open(Path("/home/unicph.domain/glj523/samtools_file_list_examples.txt")) as f:
    files = [line.strip() for line in f if line.strip()]
```

```python
# First pass: collect all metrics to build column headers
metrics = []
print('collecting metrics')

filepath_metadata = files[0].split("/")


for filepath in files:
    with open(filepath) as f:
        for line in f:
            if line.startswith("SN\t"):
                metric = line.strip().split("\t")[1].rstrip(":")
                if metric not in metrics:
                    metrics.append(metric)
```

```python
# Second pass: extract values per file
print('writing summary')
with open("summary.tsv", "w", newline="") as fout:
    writer = csv.writer(fout, delimiter="\t")
    writer.writerow(["filename"] + metrics)

    for filepath in files:
        row = {m: "" for m in metrics}
        with open(filepath) as f:
            for line in f:
                if line.startswith("SN\t"):
                    parts = line.strip().split("\t")
                    row[parts[1].rstrip(":")] = parts[2]
        writer.writerow([os.path.basename(filepath)] + [row[m] for m in metrics])
```

```python
def get_basic_stats(results) -> pd.DataFrame:
    combined_df = pd.DataFrame()
    for result in results:
        status_data = [{'Measure': f"{k} status", 'Value': v['status']} for k, v in result['modules'].items()]
        basic_stats = result['modules']['Basic Statistics']['table']['rows']
        header_data = [{'Measure': i, 'Value': j} for k, v in result['modules'].items() for i, j in v['header_values'].items()]
        df = pd.DataFrame(basic_stats + header_data + status_data)
        df = df.set_index('Measure').T.reset_index(drop=True)
        combined_df = pd.concat([combined_df, df], ignore_index=True)
    return combined_df
```
